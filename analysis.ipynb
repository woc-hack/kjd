```python
import pandas as pd

columts = ['csv', 'copy_project', 'fix_commit', 'vuln_commit', 'fix_time', 'vuln_time', 'orig_fix_commit', 'orig_project', 'orig_fix_commit_time', 'vuln_publish_time']
analysis_result = pd.read_csv('data/cve_analysis_joined.csv', sep=';', header=0, names=columts, index_col=False)
```

```python
analysis_result
```

```python
analysis_result['fix_time'] = pd.to_datetime(analysis_result["fix_time"], format="%Y-%m-%d:%H:%M:%S")
```

```python
analysis_result['vuln_time'] = pd.to_datetime(analysis_result["vuln_time"], format="%Y-%m-%d:%H:%M:%S")
```

```python
analysis_result['orig_fix_commit_time'] = pd.to_datetime(analysis_result["orig_fix_commit_time"], utc=True).dt.tz_localize(None)
```

```python
analysis_result['vuln_publish_time'] = pd.to_datetime(analysis_result["vuln_publish_time"]).dt.tz_localize(None)
```

```python
analysis_result['max_bad_time'] =  analysis_result[['vuln_publish_time', 'orig_fix_commit_time']].values.max(axis=1)
```

```python
analysis_result
```

```python
analysis_result['time-delta'] = (analysis_result['fix_time'] - analysis_result['max_bad_time']).dt.days
```

```python
analysis_result
```

```python
analysis_result.groupby('time-delta').size().plot(kind='bar')
```
